```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./SPREAD_food_networks/sn/results/networks/1.edges', sep=',', usecols = ['source', 'target', 'month'])
def dup_edges(graph_df, period = 12):
    """Duplicate each edge with new timestamps i > period+i"""    
    return pd.concat([graph_df, pd.concat([graph_df.iloc[:,[0, 1]], graph_df.iloc[:,[2]]+period], axis=1)], ignore_index=True)
df = dup_edges(df)
print(df.tail())
```

```python
G = nx.from_pandas_edgelist(df, 'source', 'target', edge_attr='month', create_using=nx.MultiDiGraph)
G = nx.convert_node_labels_to_integers(G, ordering='sorted', label_attribute = 'old_id')
list(G.edges)[0]
```

```python
edges = sorted(G.edges(data=True), key=lambda t: t[2].get('month', 1))

delta = 12
def bidirecCountAndStore(edges, delta):
    bidirec_counts = []
    for i in range(len(edges)):
        count = 0
        j = i + 1
        while j < len(edges):
            if edges[j][2]['month'] > edges[i][2]['month'] + delta:
                break
            if edges[j][0] == edges[i][1] and edges[j][1] == edges[i][0]:
                count += 1
            j+=1
        edges[i][2]['bidir_count'] = count
        bidirec_counts.append(count)
    return edges
edges_bi = bidirecCountAndStore(edges, delta = 10)
edges_bi
```
