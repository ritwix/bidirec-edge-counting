```python
dfs = []
# read in the files just generated
for name in glob.glob('data/out_sn_new_delta_?*.csv'):
    graphsn = pd.read_csv(name)
    dfs.append(graphsn) 
# and add them all to the same dataframe for analysis
df_ = pd.concat(dfs)
#print(dfs[0].to_string())
```

```python
df = df_[['t1', 'count', 'delta']].groupby(['t1', 'delta']).sum().reset_index()
sns.set(rc={'figure.figsize':(12,9)})
sns.lineplot(data=df, x='t1', y='count', hue='delta', legend='full')
```
